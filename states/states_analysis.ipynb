```python
import numpy as np
import pandas as pd
```

```python
states_df = pd.read_csv('states.csv')
states_df.set_index('state_name',inplace=True)
states_df.sort_values('water_quality_ranking')
states_df['density'] = states_df['population']/states_df['area_in_sq_mi']
states_df['college_educated'] = (states_df['population'] * states_df['college_educated_in_percentage']) / 100
states_df['industrial_toxins_lbs'] = states_df['industrial_toxins_lbs_per_sq_mi'] * states_df['area_in_sq_mi']
states_df
```
